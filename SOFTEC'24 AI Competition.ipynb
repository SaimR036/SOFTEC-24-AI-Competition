## EfficientNet architecture Used

```python
def efficientNet():

    """
    Constructs a EffNet model with customizable depth.
    """
    base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(1004,1004,3))

    split_index = int(len(base_model.layers)*0.5)

# Freeze the first half of the layers
    for layer in base_model.layers[:split_index]:
        layer.trainable = False

    x = base_model.output

    x = Conv2D(64, (7, 7), strides=(2, 2), padding='same', kernel_regularizer=l2(0.01))(x)
    x= layers.Dropout(0.5)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='same')(x)
    x = BatchNormalization()(x)
    x = GlobalAveragePooling2D()(x)
    x=Dense(64, activation='relu', kernel_regularizer=l2(0.001))(x)
    predictions = Dense(1, activation='sigmoid')(x)

# Create the final model
    model = Model(inputs=base_model.input, outputs=predictions)
    return model


model = efficientNet()
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_delta=0.001, mode='min', verbose=1)
```
